```python
import pandas as pd

data = pd.read_csv("contacts.csv")
data.head()
```

```python
for index, row in data.iterrows():
    phone = row["номер"]
    message = row["сообщение"]

    print("DRY-RUN")
    print("Номер:", phone)
    print("Сообщение:", message)
    print("Статус: имитация отправки")
    print("-" * 30)
```

```python
with open("send_log.txt", "w", encoding="utf-8") as log:
    for index, row in data.iterrows():
        phone = row["номер"]
        message = row["сообщение"]

        log.write(
            f"[DRY-RUN] Номер: {phone} | Сообщение: {message}\n"
        )

print("Лог успешно сохранён!")
```

```python
for index, row in data.iterrows():
    phone = str(row["номер"])

    if not phone.startswith("+"):
        print(f"Ошибка: неверный номер {phone}")
        continue

    print(f"{phone} прошёл проверку")
```
